```python
import pandas as pd


df = pd.read_csv('/Users/clockworkcritter/Library/CloudStorage/OneDrive-UniversityofMaryland/Fall 2025/INST462/2022BFRSSData (1).csv')
df.head()
smalldataset = df[["deaf","blind","decide","diffwalk","diffdres","diffalon","checkup1","exerany2"]].copy()
smalldataset.head()
cleandataset = smalldataset.dropna()
def check_disability(row):
    """Assigns 1 if any disability column is 1, otherwise 0."""
    if (row['deaf'] == 1) or \
       (row['blind'] == 1) or \
       (row['decide'] == 1) or \
       (row['diffwalk'] == 1) or \
       (row['diffdres'] == 1) or \
       (row['diffalon'] == 1):
        return 1
    else:
        return 0
# Apply the function across rows (axis=1)
cleandataset['Disabled_Status'] = df.apply(check_disability, axis=1)
cleandataset.to_csv('/Users/clockworkcritter/Library/CloudStorage/OneDrive-UniversityofMaryland/Fall 2025/INST462/2022BFRSSDataClean.csv')

```
